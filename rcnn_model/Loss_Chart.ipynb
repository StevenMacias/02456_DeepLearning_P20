```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(15,6))
# make a plot
ax.plot(df["1.1173"], color="blue")
# set x-axis label
ax.set_xlabel("Iteration",fontsize=14)
# set y-axis label
ax.set_ylabel("Loss",color="blue",fontsize=14)
ax.set_title("Loss and Adaptative Learning Rate Evolution Over 6 Epochs")
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df["0.000021"],color="red")
ax2.set_ylabel("Learning Rate",color="red",fontsize=14)
plt.show()
```
